```python
from pandas import read_csv
from pandas import DataFrame
import pandas as pd
from pandas import drop
```

```python
march = read_csv("FlairRatioTestDemocracyMarch2020.csv", index_col=0)
feb = read_csv("FlairRatioTestDemocracyFeb2020.csv", index_col=0)
print(march)
```

```python
def csvSplitter(pooper):
    numDays = pooper.shape[0]
    df = pd.DataFrame(data=pooper)
    print(numDays)
    totals = []
    positives = []
    negatives = []
    for i in range(numDays):
        stringArr = df.iloc[i, 1]
        splitness = stringArr.split(",")
        tot = splitness[0]
        tot = tot[1:]
        tot = float(tot)
        pos = splitness[1]
        pos = pos[1:]
        pos = float(pos)
        neg = splitness[2]
        neg = neg[1:-1]
        neg = float(neg)
        totals.append(tot)
        positives.append(pos)
        negatives.append(neg)
        
    newDF = DataFrame(data=df.iloc[:,0])
    newDF["Totals"] = totals
    newDF["Positives"] = positives
    newDF["Negatives"] = negatives
    return newDF
        
```

```python
print(march)

marchCSV = csvSplitter(march)

print(marchCSV)
print(" ")

print(feb)
febCSV = csvSplitter(feb)
print(febCSV)
```
