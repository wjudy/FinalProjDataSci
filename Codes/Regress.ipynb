```python
febTots = {0:f18tot, 1:f19tot, 2:f20tot, 3:f21tot, 4:f22tot}
febPos = {0:f18pos, 1:f19pos, 2:f20pos, 3:f21pos, 4:f22pos}
febNeg = {0:f18neg, 1:f19neg, 2:f20neg, 3:f21neg, 4:f22neg}

febCorrTots = np.ones((5,5))
for i in range(5):
    for j in range(5):
        a = febTots[i]
        b = febTots[j]
        febCorrTots[i,j] = abs((np.corrcoef(a, b))[1,0])
        
print("FEBRUARY")
print("Totals:")        
print(febCorrTots)


febCorrPos = np.ones((5,5))
for i in range(5):
    for j in range(5):
        a = febPos[i]
        b = febPos[j]
        febCorrPos[i,j] = abs((np.corrcoef(a, b))[1,0])

print("Positives: ")
print(febCorrPos)


febCorrNeg = np.ones((5,5))
for i in range(5):
    for j in range(5):
        a = febNeg[i]
        b = febNeg[j]
        febCorrNeg[i,j] = abs((np.corrcoef(a, b))[1,0])

print("Negatives: ")        
print(febCorrNeg)
print(" ")
print(" ")
avgCorrTots = np.ones((1,5))
for i in range(5):
    sum = 0
    for j in range(5):
        sum += febCorrTots[i, j]
    avgCorrTots[0,i] = sum / 5
    
print("Total Averages")    
print(avgCorrTots)
print(" ")

avgCorrTots = np.ones((1,5))
for i in range(5):
    sum = 0
    for j in range(5):
        sum += febCorrPos[i, j]
    avgCorrTots[0,i] = sum / 5
    
print("Positive Averages")    
print(avgCorrTots)
print(" ")

avgCorrTots = np.ones((1,5))
for i in range(5):
    sum = 0
    for j in range(5):
        sum += febCorrNeg[i, j]
    avgCorrTots[0,i] = sum / 5
    
print("Positive Averages")    
print(avgCorrTots)
```
