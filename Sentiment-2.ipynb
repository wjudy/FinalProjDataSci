```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer

```

```python
sid = SentimentIntensityAnalyzer()
counter = 0

overallScoreList = []
negList = []
neuList = []
posList = []

columns = ['neg', 'neu', 'pos', 'compound']


for i, tweet in enumerate(sntwitter.TwitterSearchScraper('Democracy since:2019-03-01 until:2019-03-02').get_items()):
        if counter == 50:
            break
        if not tweet.inReplyToUser and not ("@" in str(tweet.content)) and str(tweet.lang) == 'en':
            counter += 1
            stringa = str(tweet.content)
            sentDict = sid.polarity_scores(stringa)
            
            print(sentDict)
            overallScoreList.append(sentDict['compound'])
            negList.append(sentDict['neg'])
            neuList.append(sentDict['neu'])
            posList.append(sentDict['pos'])

            
dfTransfer = {'neg': negList, 'neu':neuList, 'pos':posList, 'compound':overallScoreList}

df = pd.DataFrame(dfTransfer)

print(df)


```
