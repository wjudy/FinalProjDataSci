```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import nltk
#nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Create month of March
daysMarch2019 = ["2019-03-01", "2019-03-02", "2019-03-03", "2019-03-04", "2019-03-05", "2019-03-06", "2019-03-07", "2019-03-08", "2019-03-09", "2019-03-10", "2019-03-11", "2019-03-12", "2019-03-13", "2019-03-14", "2019-03-15", "2019-03-16", "2019-03-17", "2019-03-18", "2019-03-19", "2019-03-20", "2019-03-21", "2019-03-22", "2019-03-23", "2019-03-24", "2019-03-25", "2019-03-26", "2019-03-27", "2019-03-28", "2019-03-29", "2019-03-30", "2019-03-31", "2019-04-01"]
march_2019_dict = {}

sid = SentimentIntensityAnalyzer()

for counter in range(31):
    print(counter)
    queryString = "Democracy since:" + daysMarch2019[counter] + " " + "until:" + daysMarch2019[counter+1]
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(queryString).get_items()):
        negList = []
        neuList = []
        posList = []
        overallScoreList = []
        if i == 100:
            break
        if not tweet.inReplyToUser and not ("@" in str(tweet.content)) and str(tweet.lang) == 'en':
            stringa = str(tweet.content)
            sentDict = sid.polarity_scores(stringa)
            #print(sentDict)
            #print(tweet.content)
            overallScoreList.append(sentDict['compound'])
            negList.append(sentDict['neg'])
            neuList.append(sentDict['neu'])
            posList.append(sentDict['pos'])
            dictTransfer = {'neg': negList, 'neu':neuList, 'pos':posList, 'compound':overallScoreList}
            df = pd.DataFrame(dictTransfer)

            avgNeg = sum(df['neg']) / len(df['neg'])
            avgNeu = sum(df['neu']) / len(df['neu'])
            avgPos = sum(df['pos']) / len(df['pos'])
            avgCom = sum(df['compound']) / len(df['compound'])
            print(counter)
            march_2019_dict[counter + 1] = [avgNeg, avgNeu, avgPos, avgCom]
```

```python
print(march_2019_dict)

plt.figure() 

fig, ax = plt.subplots()

dayLabels = list(march_2019_dict.keys())

neg_means = []
neu_means = []
pos_means = []
com_means = []

for i in range(31):
    current = march_2019_dict[i+1]
    neg_means.append(current[0])
    neu_means.append(current[1])
    pos_means.append(current[2])
    com_means.append(current[3])
    
    
print(dayLabels)
print(neg_means)
print(neu_means)
print(pos_means)
print(com_means)

x = np.arange(len(dayLabels))

width = 0.25

#rects1 = ax.bar(x - width/2, neg_means, width, label='Neg')
#rects2 = ax.bar(x - width/4, neu_means, width, label='Neu')
#rects3 = ax.bar(x + width/4, pos_means, width, label='Pos')
rects4 = ax.bar(x - width/2, com_means, width, label='Com')

ax.set_ylabel("avg sentiment")
ax.set_xlabel("Days of month")

ax.set_title("March 2019 compound sentiment analysis")

#ax.bar_label(rects1, padding=3)
#ax.bar_label(rects2, padding=3)
#ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)

fig.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Info from:
# https://matplotlib.org/3.5.0/gallery/lines_bars_and_markers/horizontal_barchart_distribution.html#sphx-glr-gallery-lines-bars-and-markers-horizontal-barchart-distribution-py

category_names = ['Negative', 'Neutral',
                  'Positive']

results = march_2019_dict


def survey(results, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(10, 10))
    ax.invert_yaxis()
    #ax.xaxis.set_visible(False)
    #ax.set_xlim(0, np.sum(data, axis=1).max())
    ax.set_xlim(0, 1)

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        r, g, b, _ = color
        #text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        #ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 0.96),
              loc='lower left', fontsize='small')
    ax.set_xlabel("Percentages of Sentiment")
    ax.set_ylabel("Days of Month")
    ax.set_title("Sentiment Analysis for March of 2019")
    
    return fig, ax


survey(results, category_names)
plt.show()
```
